```python
a=int(input("enter your age : "))

# if elif else ladder

if(a>=18):                     
    print("you are above the age of consent")
    print("good for you ")

 #when "if" statement is fulfilled the "elif" and "else" statement is ignored

elif(a<0):
    print("you are entering negative age")

elif(a==0):
    print("you are entering 0 , which is not valid")

else:
    print("you are below the age of consent")

print("end of program")
```

```python
# if elif else ladder

a=int(input("enter your age : "))

#if one condition is matched or fulfilled,the other conditions will be ignored . 

if(a>=18):                     
    print("you are above the age of consent")
    print("good for you ")


elif(a<0):
    print("you are entering negative age")

elif(a==0):
    print("you are entering 0 , which is not valid")


else:
    print("you are below the age of consent")

print("end of program")
```
