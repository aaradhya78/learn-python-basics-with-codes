```python
# for count 
a=(1,7,2,3)
print(a.count(1))      #a.count(1) will return number of times 1 occurs in 'a' 
```

```python
# for finding index 
a=(1,7,2,3)
print(a.index(1))

# a.index() will return the index of first occurence of 1 in 'a'
```

```python
# for length of tuple
a=(1,7,2,3)

print(len(a))
```

```python
# for largest value

a=(1,7,2,3,8,5)

print(max(a))

# will give the largest value present in the tuple 
```

```python
# for smallest value 

a=(1,7,2,3,8,5)

print(min(a))

# will give the smallest value present in the tuple 
```

```python
# for sum of numeric tuple 

a=(1,7,2,3,8,5)

print(sum(a))

# sum of all numeric terms 
```

```python
# for sorting 

a=(1,7,2,3,8,5)

print(sorted(a))

# this will returns a sorted list
```

```python
# for converting it into tuple 

a=(1,7,2,3,8,5)

print(tuple(a))
```

```python
# for concatenation 
# tuple concatenation using "+" operator 
# combining two tuples

tuple1=(1,2,3)
tuple2=(4,5,6)

concatenated = tuple1 + tuple2 

print(concatenated)
```

```python
# for repetition
# tuples can be repeated using the "*" operator

a=(1,2,3)
repeated = a*3

print(repeated)
```

```python
# for membership
# we can check if the item exists in a tuple , using "in" keyword

a=(1,2,3)
print( 2 in a)
print( 4 in a)

# give the output in true/false
```

```python
# for slicing
# tuple support slicing to create a new tuple from a subset of the original

my_tuple=(1,2,3,4,5)
sliced = my_tuple[1:4]

print(sliced)
```
