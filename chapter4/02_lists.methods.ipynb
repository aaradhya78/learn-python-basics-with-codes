```python
#for sorting L1.sort

L1=[1,8,7,2,21,15]
L1.sort()
print(L1)
```

```python
# for reverse L1.reverse()

L1=[1,8,7,2,21,15]
L1.reverse()
print(L1)
```

```python
# for append  L1.append() 
#add element at the end 

L1=[1,8,7,2,21,15]
L1.append(10)
print(L1)
```

```python
# for insert  L1.insert( , )


L1=[1,8,7,2,21,15]
L1.insert(3,8)          # this will add 8 at the 3rd index
print(L1)
```

```python
# for pop   L1.pop( )
#will delete element at the index and return its value
 
L1=[1,8,7,2,21,15]
L1.pop(2)
print(L1)
```

```python
# for remove   L1.remove()
# will remove/delete that element from the list

L1=[1,8,7,2,21,15]
L1.remove(2)
print(L1)
```
