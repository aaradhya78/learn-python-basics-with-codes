```python
s={1,8,2,3}

s.union({8,11,4})

# union will combines all the elements of the sets
```

```python
s1={2,4,7,6,9}

s2={3,7,9,5,1}

print(s1.union(s2))
```

```python
s={1,8,2,3}

print(s.intersection({8,11}))

#returns the set of the same element 
```

```python
s1={2,4,7,6,9}

s2={3,7,9,5,1}

print(s1.intersection(s2))
```

```python
# using issubset()
s1={2,4,7,6,9}

print({4,2}.issubset(s1))
```

```python
# using issuperset()
s1={3,7,9,5,1}

print(s1.issuperset({1,3}))
```
