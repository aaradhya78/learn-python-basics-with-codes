```python
marks = {
    "harry" : 80,
    "raju": 39,
    "rohan": 60,
    0   :  "harry"
}

print(marks.items())  #returns a list of (keys,values)tuples

print(marks.keys())  #the right side of the marks are keys

print(marks.values())     #the left side of the marks are values

```

```python
marks = {
    "harry" : 80,
    "raju": 39,
    "rohan": 60,
    0   :  "harry"
}

# update dictionary with supplied key-values pairs

marks.update({"harry":99})
marks.update({"renu":100})       
print(marks)
```

```python
marks = {
    "harry" : 80,
    "raju": 39,
    "rohan": 60,
    0   :  "harry"
}

print(marks.get("harry"))

print(marks.get("shreya"))  
#there is no "shreya" in this dictionary , the output will be none
```
