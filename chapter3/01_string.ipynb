```python
# introduction to string

#this is string slicing

name = "harry"

a = name [0:3]   #start from index 0 all the way till 3 (excluding 3) "HAR"
print(a)

character1=name[1]
print (character1)

b=name[0]
print(b)
```

```python
# for negetive slicing

name = "harry"

print(name[-4:-1])
print(name[1:4])
```

```python
# for blank space

name = "harry"

print(name[:4]) # is same as [0:4]
print(name[1:]) # is same as [1:5]
print(name[1:5]) # is same as [1:5]
```

```python
# slicing with skip values


# example 1
word= "amazing"
word[1:6:2]

# m-a-z-i-n-g 
# m---z---n 
# "m z n"


# example2
a="0123456789"
a[1:7:3]

# 1-2-3-4-5-6
# 1-!-!-4-!-!-x
# '1 4'

```

```python
# other slicing techniques

word= "amazing"
word[0:7]  #output will be [0:7] "amazing"

word[0:] #output will be [0:7] "amazing"
```
