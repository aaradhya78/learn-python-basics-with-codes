```python
# for new line \n

a= "i am aaradhya sharma ,\n nice to meet \"you\""
print (a)
```

```python
# for tab \t (for space)

a= "i am aaradhya sharma \t,nice to meet \"you\""
print (a)
```

```python
# for single qoute ( \'     \')

a= "i am aaradhya sharma ,nice to meet \'you\'"
print (a)
```

```python
# for double qoute ( \"     \")

a= "i am aaradhya sharma ,nice to meet \"you\""
print (a)
```

```python
# for backslash

a= "i am aaradhya sharma ,\\ nice to meet \"you\""
print (a)
```

```python
# for carriage return \r

a= "i am aaradhya sharma ,\r nice to meet \'you\'"
print (a)
```

```python
# for backspace \b

a= "i am aaradhya sharma ,\b nice to meet \'you\'"
print (a)
```
