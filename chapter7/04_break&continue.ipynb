```python
# the break statement 

for i in range(0,80):

    if (i == 30) :
        break  # it will exit the loop right now

    print(i)

```

```python
# the continue statement 

for i in range(0,80):

    if (i == 30) :
        continue     # it will skip the iteration in the loop , i == iteration

    print(i)
```
